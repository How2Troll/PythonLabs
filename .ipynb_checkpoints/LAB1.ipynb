```python
import os
import re
import random
```

```python
#ZAD 1
def licz_pliki_w_katalogu(katalog):
    try:
        pliki = [f for f in os.listdir(katalog) if os.path.isfile(os.path.join(katalog, f))]
        return len(pliki)
    except Exception as e:
        print(f"Wystąpił błąd: {e}")
        return 0

katalog = "C:\\Windows"
ilosc_plikow = licz_pliki_w_katalogu(katalog)
print(f"W katalogu {katalog} jest {ilosc_plikow} plików.")
```

```python
#ZAD 2

def list_files(directory):
    try:
        # Check if the given path is a directory
        if os.path.isdir(directory):
            for name in os.listdir(directory):
                full_path = os.path.join(directory, name)
                if os.path.isdir(full_path):
                    list_files(full_path)  # Recursive call for directories
                else:
                    print(full_path)
        else:
            print(f"{directory} is not a directory!")
    except Exception as e:
        print(f"An error occurred: {e}")

directory = r"C:\Users\igosie\Documents\Repo"
list_files(directory)

```

```python
#ZAD3
def remove_words_from_file(filename, words_to_remove):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            text = file.read()

        for word in words_to_remove:
            text = text.replace(word, "")

        with open(filename, 'w', encoding='utf-8') as file:
            file.write(text)

        print("Specified words have been removed from the file.")

    except Exception as e:
        print(f"An error occurred: {e}")

filename = "test.txt"
words = "test,uwaga"
words_to_remove = [word.strip() for word in words.split(",")]

remove_words_from_file(filename, words_to_remove)
```

```python
#ZAD 4

def replace_words_in_file(filename, replacement_dict):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            text = file.read()

        for original_word, replacement_word in replacement_dict.items():
            text = text.replace(original_word, replacement_word)

        with open(filename, 'w', encoding='utf-8') as file:
            file.write(text)

        print("Words have been replaced in the file.")

    except Exception as e:
        print(f"An error occurred: {e}")

filename = "test.txt"
replacement_dict = {'uwaga': 'UwAgA', 'yua': 'ayu'}

replace_words_in_file(filename, replacement_dict)
```

```python
# ZAD 5

def remove_words_from_file(filename, words_to_remove):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            text = file.read()

        for word in words_to_remove:
            pattern = r'\b' + re.escape(word) + r'\b'
            text = re.sub(pattern, '', text)

        with open(filename, 'w', encoding='utf-8') as file:
            file.write(text)

        print("Specified words have been removed from the file.")

    except Exception as e:
        print(f"An error occurred: {e}")

def replace_words_in_file(filename, replacement_dict):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            text = file.read()

        for original_word, replacement_word in replacement_dict.items():
            pattern = r'\b' + re.escape(original_word) + r'\b'
            text = re.sub(pattern, replacement_word, text)

        with open(filename, 'w', encoding='utf-8') as file:
            file.write(text)

        print("Words have been replaced in the file.")

    except Exception as e:
        print(f"An error occurred: {e}")

filename = "test.txt"
words = "test,uwaga"
remove_words_from_file(filename, [word.strip() for word in words])

replacement_dict = {'uwaga': 'UwAgA', 'yua': 'ayu'}
replace_words_in_file(filename, replacement_dict)
```

```python
#ZAD6

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

N = 100
numbers = [random.randint(1, 1000) for _ in range(N)]
print(f"Generated numbers: {numbers}")

# Sort using bubble sort
sorted_by_bubble = bubble_sort(numbers.copy())
print(f"Sorted by Bubble Sort: {sorted_by_bubble}")

# Sort using insertion sort
sorted_by_insertion = insertion_sort(numbers.copy())
print(f"Sorted by Insertion Sort: {sorted_by_insertion}")

# Verify using Python's built-in sort
verified_sorted = sorted(numbers)
assert sorted_by_bubble == verified_sorted, "Bubble Sort failed verification!"
assert sorted_by_insertion == verified_sorted, "Insertion Sort failed verification!"
print("Both sorting methods passed verification!")
```

```python
class BinaryTreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

    def insert_left(self, node):
        """Dodaj węzeł jako lewe dziecko."""
        if self.left is None:
            self.left = node
        else:
            node.left = self.left
            self.left = node

    def insert_right(self, node):
        """Dodaj węzeł jako prawe dziecko."""
        if self.right is None:
            self.right = node
        else:
            node.right = self.right
            self.right = node

    def traverse(self):
        """Przechodzi przez drzewo i zwraca wszystkie wartości węzłów w liście."""
        nodes = [self.value]
        if self.left:
            nodes.extend(self.left.traverse())
        if self.right:
            nodes.extend(self.right.traverse())
        return nodes

    def __str__(self, level=0):
        """Zwraca tekstową reprezentację drzewa."""
        ret = "\t" * level + repr(self.value) + "\n"
        if self.left:
            ret += self.left.__str__(level + 1)
        if self.right:
            ret += self.right.__str__(level + 1)
        return ret

# Przykład użycia:
root = BinaryTreeNode("Korzeń")

left1 = BinaryTreeNode("L1")
right1 = BinaryTreeNode("P1")

left2_left = BinaryTreeNode("L2-L")
left2_right = BinaryTreeNode("L2-P")
right2_left = BinaryTreeNode("P2-L")
right2_right = BinaryTreeNode("P2-P")

left3_LL = BinaryTreeNode("L3-L-L")
left3_LR = BinaryTreeNode("L3-L-R")
left3_RL = BinaryTreeNode("L3-R-L")
left3_RR = BinaryTreeNode("L3-R-R")
right3_LL = BinaryTreeNode("R3-L-L")
right3_LR = BinaryTreeNode("R3-L-R")
right3_RL = BinaryTreeNode("R3-R-L")
right3_RR = BinaryTreeNode("R3-R-R")

root.insert_left(left1)
root.insert_right(right1)

left1.insert_left(left2_left)
left1.insert_right(left2_right)
right1.insert_left(right2_left)
right1.insert_right(right2_right)

left2_left.insert_left(left3_LL)
left2_left.insert_left(left3_LR)
left2_right.insert_left(left3_RL)
left2_right.insert_left(left3_RR)
right2_left.insert_left(right3_LL)
right2_left.insert_left(right3_LR)
right2_right.insert_left(right3_RL)
right2_right.insert_left(right3_RR)

print("Przejście po drzewie:", root.traverse())
print("\nStruktura drzewa:\n", root)
```

```python
import unittest

class TestBinaryTreeNode(unittest.TestCase):

    def setUp(self):
        """Ustawienie początkowe przed każdym testem."""
        self.root = BinaryTreeNode("Korzeń")
        self.left1 = BinaryTreeNode("L1")
        self.right1 = BinaryTreeNode("P1")

        self.root.insert_left(self.left1)
        self.root.insert_right(self.right1)

    def test_insert_left(self):
        """Testuje dodawanie lewego dziecka."""
        self.assertEqual(self.root.left, self.left1)

    def test_insert_right(self):
        """Testuje dodawanie prawego dziecka."""
        self.assertEqual(self.root.right, self.right1)

    def test_traverse(self):
        """Testuje przechodzenie przez drzewo."""
        expected_traversal = ["Korzeń", "L1", "P1"]
        self.assertEqual(self.root.traverse(), expected_traversal)

    def test_str_representation(self):
        """Testuje tekstową reprezentację drzewa."""
        expected_str = "'Korzeń'\n\t'L1'\n\t'P1'\n"
        self.assertEqual(str(self.root), expected_str)

suite = unittest.TestLoader().loadTestsFromTestCase(TestBinaryTreeNode)
unittest.TextTestRunner().run(suite)
```
