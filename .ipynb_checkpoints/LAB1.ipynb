```python
import os
```

```python
#ZAD 1
def licz_pliki_w_katalogu(katalog):
    try:
        pliki = [f for f in os.listdir(katalog) if os.path.isfile(os.path.join(katalog, f))]
        return len(pliki)
    except Exception as e:
        print(f"Wystąpił błąd: {e}")
        return 0

katalog = "C:\\Windows"
ilosc_plikow = licz_pliki_w_katalogu(katalog)
print(f"W katalogu {katalog} jest {ilosc_plikow} plików.")
```

```python
#ZAD 2

def list_files(directory):
    try:
        # Check if the given path is a directory
        if os.path.isdir(directory):
            for name in os.listdir(directory):
                full_path = os.path.join(directory, name)
                if os.path.isdir(full_path):
                    list_files(full_path)  # Recursive call for directories
                else:
                    print(full_path)
        else:
            print(f"{directory} is not a directory!")
    except Exception as e:
        print(f"An error occurred: {e}")

directory = r"C:\Users\igosie\Documents\Repo"
list_files(directory)

```

```python
#ZAD3
def remove_words_from_file(filename, words_to_remove):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            text = file.read()

        for word in words_to_remove:
            text = text.replace(word, "")

        with open(filename, 'w', encoding='utf-8') as file:
            file.write(text)

        print("Specified words have been removed from the file.")

    except Exception as e:
        print(f"An error occurred: {e}")

filename = "test.txt"
words = "test,uwaga"
words_to_remove = [word.strip() for word in words.split(",")]

remove_words_from_file(filename, words_to_remove)
```

```python
#ZAD 4

def replace_words_in_file(filename, replacement_dict):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            text = file.read()

        for original_word, replacement_word in replacement_dict.items():
            text = text.replace(original_word, replacement_word)

        with open(filename, 'w', encoding='utf-8') as file:
            file.write(text)

        print("Words have been replaced in the file.")

    except Exception as e:
        print(f"An error occurred: {e}")

filename = "test.txt"
replacement_dict = {'uwaga': 'UwAgA', 'yua': 'ayu'}

replace_words_in_file(filename, replacement_dict)
```

```python
# ZAD 5

def remove_words_from_file(filename, words_to_remove):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            text = file.read()

        for word in words_to_remove:
            pattern = r'\b' + re.escape(word) + r'\b'
            text = re.sub(pattern, '', text)

        with open(filename, 'w', encoding='utf-8') as file:
            file.write(text)

        print("Specified words have been removed from the file.")

    except Exception as e:
        print(f"An error occurred: {e}")


def replace_words_in_file(filename, replacement_dict):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            text = file.read()

        for original_word, replacement_word in replacement_dict.items():
            pattern = r'\b' + re.escape(original_word) + r'\b'
            text = re.sub(pattern, replacement_word, text)

        with open(filename, 'w', encoding='utf-8') as file:
            file.write(text)

        print("Words have been replaced in the file.")

    except Exception as e:
        print(f"An error occurred: {e}")

filename = "test.txt"
words = "test,uwaga"
remove_words_from_file(filename, [word.strip() for word in words])

replacement_dict = {'uwaga': 'UwAgA', 'yua': 'ayu'}
replace_words_in_file(filename, replacement_dict)
```
